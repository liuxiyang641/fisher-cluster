```python
import pandas as pd
import numpy as np
import math
```

```python
origin_data = pd.read_excel("./fisher.xls")
```

```python
origin_data
```

```python
# 下雨的数据
g1_data = origin_data.iloc[:10]
g1_data
```

```python
# 不下雨的数据
g0_data = origin_data.iloc[10:]
g0_data
```

```python
x0_mean = pd.DataFrame(g0_data.iloc[:,:2].mean(axis=0))
x0_mean
```

```python
x1_mean = pd.DataFrame(g1_data.iloc[:,:2].mean(axis=0))
x1_mean
```

```python
x_mean = pd.DataFrame(origin_data.iloc[:,:2].mean(axis=0))
x_mean
```

```python
B = (x0_mean - x_mean) @ (x0_mean - x_mean).T + (x1_mean - x_mean) @ (x1_mean - x_mean).T
B
```

```python
E = g0_data.iloc[:, :2].cov() + g1_data.iloc[:, :2].cov()
E
```

```python
E_inverse = pd.DataFrame(np.linalg.inv(E.values), E.columns, E.index)
E_inverse
```

```python
E_inverse @ E
```

```python
target_matrix = E_inverse @ B
target_matrix
```

```python
e_values, e_vectors = np.linalg.eig(target_matrix)
e_values
```

```python
e_vectors
```

```python
alpha = pd.DataFrame(e_vectors[:, 1], E.columns,)
alpha
```

```python
bias_value = alpha.T @ E @ alpha
bias_value
```

```python
alpha = alpha / math.sqrt(bias_value.iloc[0, 0])
```

```python
alpha
```

```python
alpha.T @ E @ alpha
```

```python
# 开始计算误回报率
res_g0 = []
res_g1 = []
for idx in origin_data.index:
    res_g0.append(
        (alpha.T @ (origin_data.iloc[[idx], :2].T.values - x0_mean)).iloc[0, 0]
    )
    res_g1.append(
        (alpha.T @ (origin_data.iloc[[idx], :2].T.values - x1_mean)).iloc[0, 0]
    )
```

```python
origin_data["G0 val"] = res_g0
origin_data["G1 val"] = res_g1
```

```python
origin_data["predict"] = origin_data.apply(
    lambda x: 0 if abs(x["G0 val"]) < abs(x["G1 val"]) else 1, axis=1
)
origin_data
```

```python
origin_data.to_excel("./fisher_clu_res.xlsx")
```
